### Encodage des valeurs d'entraînements et de test

L'encodage a pour but de rendre plus cohérent les valeurs utilisés pour l'entraînement

```python
# Séparation des caractéristiques (X) et de la cible (y)
X = df_calc_population.drop(["Code_commune"], axis=1)  # Sélectionnez les colonnes correspondant aux caractéristiques
y = df3['Population totale']  # Sélectionnez la colonne de la variable cible

ohe = OneHotEncoder(handle_unknown='ignore')

encoded_df = pd.DataFrame(ohe.fit_transform(X[['Code_commune']]).toarray())
encoded_df.columns = ohe.get_feature_names_out()
encoded_df = encoded_df.astype("int")
encoded_df
```

### Une fois l'encodage terminée, je peux créer le dataframe que je vais utiliser pour l'entraînement.

```python
dataframe = X.join(encoded_df)
dataframe
```

```python
X_aleger = X.drop(["Code_commune"], axis=1, inplace=True)
df_join = X.join(encoded_df)
df_join
```

### Mise en place

Mise en place des données d'entraînement "X_train, y_train" et des données de test "X_test, y_test"

X étant les données entrée (input) soit : Code_commune_401 à Code_commune_424 + Année recensement

y étant les données de test (output) soit : Population totale

```python

# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(df_join, y, test_size=0.1, random_state=42)  # 0.2 => 80% entraînement, 20% test
model = LinearRegression()
```

### Entraînement

```python
# Entraînement du modèle
model.fit(X_train, y_train)
```

### Score

un score de "0.59.." se lis 59% de précision

```python
score = model.score(X_test, y_test)
print(f"Score du modèle sur les données de test : {score}")
```
